模型介紹：
這是一個可以生成不同3D椅子模型的GAN模型，在網頁中可以選擇查看椅子的任一角度，也可以透過滑桿控制並調整椅子的 Resolution、Weirdness 和 Isosurface，打造獨一無二的椅子。

##使用GAN模型經過
剛開始使用椅子GAN模型的時候，我因為沒想到這個生成椅子的模型還能使用滑桿調整各種細節而感到新鮮，不過隨著逐漸增加生成椅子的次數，我發現這個GAN模型**雖然可以在單一類別依據需求調整椅子模型的精細程度，但是不像現在流行的生圖模型可以使用文字生成想要的椅子模型。**

另外，在生成椅子的過程中，我也有以下發現：

1. 椅子的變化不多（不以滑桿控制調整時）


    使用GAN模型隨機生成的椅子大多有種熟悉感（沒有真正的創新），而且在一輪測試生成中會重複看到同一種類的椅子多次出現

2. 椅子在構造上出現不合物理的增生或減少，有時候甚至出現無法生成椅子的情形


    如果以現實椅子的標準來看，隨機生成椅子時有機率出現椅子無法乘坐的情形，例如：椅子缺腳、椅子造型過於簡略、椅子不合物理規則

3. 生成椅子的穩定性不高


    3.1 同一張椅子
    同一張椅子的椅腳不統一，比如椅子3中有4隻椅腳，其中有3隻是方形構造，有1隻是尖腳

    3.2 不同張椅子
    隨機生成椅子的品質無法穩定，無法保證每一次生成椅子時的品質

基於上述發現，我推測GAN現在比較少被人使用是因為其**穩定性**及**變化性**較低所致。
