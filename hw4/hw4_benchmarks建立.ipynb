###這次使用的 benchmark 的內容（角色設定），主要涵蓋以下幾個方面：

1. 深化角色細節：

    要求 AI 根據已提供的角色描述（大航海時代的男性海盜角色），進一步延伸並擴充角色的詳細資料，例如：飲食偏好、成長經歷、人生態度與信念、行為風格與喜好、說話與動作習慣、人生理想與目標

2. 補充背景故事：
    
    要求 AI 根據已提供的角色資料（時代背景、性格、家庭背景、成長環境、人生轉折點、職業發展），補充可前後呼應的細節設定以達到充分展現角色的核心概念

3. 合理的世界觀設定：
    
    已提供的角色描述中，明確指出角色身處**大航海時代（地理大發現，約15~17世紀）**且其職業為**海盜**。因此設定角色所在的世界觀應適度參考大航海時代的海盜生活情形，並且避免人物設定中出現不合時宜的內容。

##三、測試感想

1. **ChatGPT 推理模型**（思考時間花了43秒）：

    ChatGPT 的回答在設定角色的經歷時，大致不脫離已有的角色資料。不過，在性格、飲食喜好和說話風格等方面，可以看到 ChatGPT 除了提綱挈領列出相關項目，還有**逐項詳細解釋內容**——這部分有助於設定時更理解角色的動機。
    

    舉例：飲食偏好

    樸實實用的食物：
    身為海邊長大的孩子，他習慣了粗糙卻能夠解饞的食物。喜歡鹹魚、臘肉和簡單的麵包，這些食物不僅富含熱量，也能在長時間航行中提供所需的能量。
    
    海鮮的懷舊情懷：
    他尤其偏愛當地捕撈的新鮮海鮮，這不僅滿足了生理需求，更勾起了對兒時在海邊自由奔跑時光的美好回憶。
    
    酒類的慰藉：
    在繁重的工作與精神壓力之下，他偶爾會用一杯帶著苦澀卻又溫暖的朗姆酒來舒緩情緒，這也成了他與自己過去對話的一種方式。

2. **Grok 推理模型**（思考時間花了25秒）：

    雖然同樣使用的是推理模型， Grok 在解釋內容方面相較ChatGPT而言更為精簡，但在同一大架構下卻能條列出更多細項。

    另外， Grok 也補充了 ChatGPT 沒回答到的動作習慣並舉例說明**（展現創造性）**，還額外解釋了角色的核心概念。


    舉例：說話習慣和動作習慣

    說話簡潔冷淡：
    他不喜歡廢話，語氣常帶冷漠，例如「說完就做，別磨蹭」。這反映了他對效率的追求和對他人的距離感。

    黑色幽默：
    在熟悉的船員面前，他偶爾會說些帶刺的笑話，例如「這船要是沉了，至少魚有得吃」。

    摸下巴或刀柄：
    思考或緊張時，他會下意識摸摸下巴或握住腰間的刀柄，這是海盜生涯養成的警覺習慣。

    銳利眼神：
    他習慣用犀利的目光掃視周圍，隨時觀察潛在威脅，這是他多年海上生活的痕跡。


3. **Claude 3.7 Sonnet**：

    與 ChatGPT 、 Grok 不同， Claude 除了達到 benchmark 的基本要求，更提供格外的設定內容（角色的姓名、外號和出身）並補充角色面臨人生轉折點時的內心想法和轉變契機。

    此外， Claude 也在設定中展現出基於時代背景所做的設定，例如在設定角色的家庭背景時明確指出「**角色出生於英格蘭南部一個貧困的漁村家庭，他的家族世代以捕魚和沿海貿易維生**」。

    另一方面， Claude 補充了許多既有設定中不夠完善的內容，展現其創造性。


    舉例：轉變契機

    在一次劫掠中，傑克的船攻擊了一艘滿載家庭移民的船隻。看到船上驚恐的孩童讓他想起了自己的弟妹，這成為他決定改變人生方向的關鍵時刻。從那時起，他開始秘密策劃如何讓整個船隊轉型，如何讓自己的手下獲得正當收入，以及如何保護大家不被過去的敵人報復。

    這個轉變不是一蹴而就的，而是一個漫長且充滿風險的過程，代表著傑克內心深處一直存在的良知和希望終於得到了釋放。



4. **perplexity 預設模型**：

    與前三個LLM不同， perplexity 回答時多使用連貫式敘述，以完整的段落來描述。
    
    此外， perplexity 也在回答時同時查詢與大航海時代相關的資料，但這些資料並沒有讓 perplexity 在設定時展現出更多的創造力。
    
    在我看來， perplexity 所提供的內容有許多偏向整理已有的角色資料，而非延伸細節。
    
    有趣的是，當我仔細查看資料來源時發現 perplexity 也收集了一些**與這次 benchmark 關聯性不高且具有誤導性的網站**，像是其中一個資料來源來自維基百科中 Fate/Grand Order 遊戲的角色列表，還有一個來自大航海時代 Online 遊戲的官網。


    舉例：說話習慣和動作習慣
    艾德溫說話時常常帶有一點點的笑意，但他的眼睛卻透露著深深的悲傷。他有個習慣，就是在說重要的事情前會停頓一下，好像在衡量每個字的重量。動作上，他經常會不自覺地摸自己的左眉毛，這是他在思考或感到緊張時的習慣。

5. **Gemini 2.0 Flash**：

    Gemini 的回覆相較其他 LLM 簡短，比較適合用於快速理解角色的生平，而不能深入理解角色的背景故事。


    舉例：核心概念

    這位海盜角色是一個複雜而矛盾的個體，他的外表和內心、行為和情感，都存在著巨大的反差。

    他是一個在逆境中成長的強者，用自己的智慧和力量，在殘酷的世界中生存。

    他對人性的不信任，源自於他悲慘的童年，以及在社會底層被剝削的經驗。

    他對家人的強烈責任感，是他做任何事情的強大動力。

    他對自由的渴望，是他一直想要脫離海盜生活，並轉型成商人的原因。


##四、小結
