##3.繪製圖形-使用colab繪製

圖片由黃色和綠色的點及線條組成

註：這裡的圖可能會跑比較久（實測約30秒）



```python
# 設定Clifford吸引子的參數（改變這些值會得到完全不同的效果）
a, b, c, d = -1.4, 1.6, 1.0, 0.7
n_iter = 1000000  # 迭代次數
x = np.empty(n_iter)
y = np.empty(n_iter)
x[0], y[0] = 0.1, 0.1  # 初始值

# 迭代生成點
for i in range(n_iter - 1):
    x[i+1] = np.sin(a * y[i]) + c * np.cos(a * x[i])
    y[i+1] = np.sin(b * x[i]) + d * np.cos(b * y[i])

# 用顏色映射（根據迭代次序來上色）
colors = np.linspace(0, 1, n_iter)

plt.figure(figsize=(10,10))
plt.scatter(x, y, c=colors, s=0.1, cmap='viridis', alpha=0.3)
plt.axis('equal')
plt.axis('off')
plt.show()
```
